### Q5) Analyze which product categories are more popular among females vs male customers.

```python
# also done as gender_cat_count = Customer_Final.groupby(['prod_cat','Gender']).agg({'Gender': 'count'})

female_prod_cat = Customer_Final[ Customer_Final['Gender'] == 'F']
display('Female product category:',female_prod_cat['prod_cat'].value_counts(normalize=True,sort=False)
                                                              .reset_index(name='Female Percentage'))

male_prod_cat = Customer_Final[Customer_Final['Gender'] == 'M']
display('Male product category:',male_prod_cat['prod_cat'].value_counts(normalize=True,sort=False)
                                                          .reset_index(name = 'Male Percentage'))
```

### Q6) Which City code has the maximum customers and what was the percentage of customers from that city?

```python
# Group by the city code and get the total number of customers
count_cust_city = Customer_Final.groupby(['city_code']).agg({'cust_id': 'count'})

max_city_code = count_cust_city.sort_values('cust_id',ascending = False).head(1)

max_city_code['Customer_Percentage'] = (max_city_code.cust_id/count_cust_city.cust_id.sum())* 100

print('The maximun count of customers and its percentage are from the city code:')
display(max_city_code)



```

### Q7) Which store type sells the maximum products by value and by quantity?

```python
# Grouping the data by store type and the calculating the sum of total_amt and Qty
max_prod_value = Customer_Final.groupby(['Store_type']).agg({'total_amt' : 'sum', 'Qty' : 'sum'})

#Sorting the values in descending order and fetching the top record
print('The store type that sells the  max. products based on value and quantity is:' )
max_prod_value.sort_values(['total_amt'],ascending = False).head(1)
```

### Q8) What was the total amount earned from the "Electronics" and "Clothing" categories from Flagship Stores?

```python
# Getting the flagship records for electronics and clothing
flagship_rec = Customer_Final[(Customer_Final['Store_type'] == 'Flagship store') & 
               ((Customer_Final['prod_cat'] == 'Electronics') | (Customer_Final['prod_cat'] == 'Clothing'))]

# Calculating the total amount earned from the required records
total_earned_amt = flagship_rec['total_amt'].sum()
print('The total amount earned from the "Electronics" and "Clothing" categories from Flagship Stores is :',total_earned_amt)
```

### Q9) What was the total amount earned from "Male" customers under the "Electronics" category?

```python
# Getting the male and electronics records
male_ele_rec = Customer_Final[(Customer_Final['Gender'] == 'M') & (Customer_Final['prod_cat'] == 'Electronics')]  

# Calculating the total amount earned
male_ele_total_amt = male_ele_rec['total_amt'].sum()
print('The total amount earned from "Male" customers under the "Electronics" category is',male_ele_total_amt)
```

### Q10) How many customers have more than 10 unique transactions, after removing all transactions which have any negative amounts?


```python
# Removing the negative transactions and saving it to a new dataframe object
Customer_Final_Positive = Customer_Final[Customer_Final['total_amt'] > 0]

# Grouping by the cust id to get their total no. of transactions
no_cust_tran = Customer_Final_Positive.groupby(Customer_Final_Positive['cust_id']).agg({'transaction_id' : 'count'}) 

# Getting customers that have more that 10 transactions and then printing the no. of customers
required_cust_id =  no_cust_tran[no_cust_tran.transaction_id > 10]

print('No. of customers that have more than 10 transaction are :',required_cust_id.count())
```

### Q11) For all customers aged between 25 - 35, find out:
### a. What was the total amount spent for “Electronics” and “Books” product categories?
### b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?


```python
# Adding Age column to the Customer_Final

today = datetime.date.today()
Customer_Final['Age'] = Customer_Final['DOB'].apply(
               lambda x: today.year - x.year - 
               ((today.month, today.day) < (x.month, x.day)) 
               )
```

```python
# Getting the required customers aged between 25 and 35

req_customers = Customer_Final[Customer_Final['Age'].between(25, 35, inclusive = True)]
```

```python
# 11 a) Getting the total amount spent for “Electronics” and “Books” product categories

# getting the customers with “Electronics” and “Books” product categories
cust_ele_book = req_customers[(req_customers['prod_cat'] == 'Electronics') | (req_customers['prod_cat'] == 'Books')]

# Calculating the total amount foe them and displaying it
cust_ele_book_tot_amt = cust_ele_book['total_amt'].sum()
print('The total amount spent for “Electronics” and “Books” product categories is:',cust_ele_book_tot_amt)
```

```python
# 11 b) The total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014

cust_date_range = Customer_Final[Customer_Final['tran_date'].between('2014-01-01', '2014-03-01', inclusive = True)]

print('The total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014 is :',cust_date_range['total_amt'].sum())
```
